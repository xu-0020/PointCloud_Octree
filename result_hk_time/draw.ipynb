# HRtree Time_cost-Time_range Test
x: the time query range (point count return).<br>
y: the time cost of the query.(in microseconds)<br>
Tree Building time:  5304635ms <br>
Note: keep the bound to the bound of whole dataset.<br>
The graph basically shows a linear growth trend, indicating that the time is basically proportional to the desired number of points

```python
file_path = "FboundRVtime_HR.txt"
the_first = 1
time_list_FboundRVtime_HR = []
count_list_FboundRVtime_HR = []
the_last=1000000000
try:
    with open(file_path, "r") as file:
        for line in file:
            if(the_first):
                the_first = 0
                continue
            # Process each line here
            line = line.strip()  # Example: Print each line
            index = line.index("Time cost =") + len("Time cost =")
            words = line[index:].split()
            if float(words[0])>the_last*10.:
                continue
            time_list_FboundRVtime_HR.append(float(words[0]))
            the_last = float(words[0])
            count_list_FboundRVtime_HR.append(float(words[-2]))
except FileNotFoundError:
    print("File not found.")
# import matplotlib.pyplot as plt

# plt.plot(count_list,time_list,'s-',color = 'r',label="ATT-RLSTM")#s-:方形
# plt.xlabel("point count")#横坐标名字
# plt.ylabel("query time cost(microseconds)")#纵坐标名字
# plt.legend(loc = "best")#图例
# plt.show()
```

# Analysis
From the result, the HR-Tree has nearly the same construction with RTree, and the performance of Time-Time range has as good performance as the RTreeT, but the Time-Coordinate has a bad performance while the RTreeT still good, the question may be that we didn't has a managament of the time.<br>
HRtree: When Input the data, first split by the time, and then construct RTree for each timestamp.<br>
RTreeT: Which is a 4d RTree, considering the time as another dimansion.

```python

plt.title('Result Analysis when fixed bound change time')
plt.plot(count_list_FboundRVtime_HR, time_list_FboundRVtime_HR, color='green', label='HRTree')
plt.plot(count_list_FtimeRVbound_RT, time_list_FtimeRVbound_RT, color='red', label='RTreeT')
plt.legend() # 显示图例


plt.xlabel('Point count')
plt.ylabel('Time cost(microsecond)')
plt.show()
```
